```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
label =['kNN', 'LR','NB', 'DT', 'RF', 'SVM']
```

```python
def plot_bar_x(values, col, y, title):
    # this is for plotting purpose
    index = np.arange(len(label))
    plot=plt.bar(index, values, color=col)
#     plot[0].set_color('r')
#     plot[1].set_color('b')
#     plot[2].set_color('g')
#     plot[3].set_color('y')
#     plot[4].set_color('purple')
#     plot[4].set_color('orange')
    
    
    
    plt.xlabel('Classifiers', fontsize=5)
    plt.ylabel(y, fontsize=5)
    plt.xticks(index, label, fontsize=5, rotation=0)
    plt.title(title)
    plt.show()
```

```python
accuracy_val =[0.9580, 0.9720, 0.9370, 0.8881, 0.9300, 0.9720]
plot_bar_x(accuracy_val, 'blue', 'Accuracy', 'Accuracy of Classifiers')
```

```python
specificity_val =[0.9888, 0.9888, 0.9777, 0.9111, 0.9111, 1.0]
plot_bar_x(specificity_val, 'orange', 'Specificity', 'Specificity of Classifiers')
```

```python
sensitivity_val =[0.9056, 0.9434, 0.8680, 0.8490, 0.9056, 0.9245]
plot_bar_x(specificity_val, 'green', 'Sensitivity', 'Sensitivity of Classifiers')
```

```python
precision_val =[0.9795, 0.9803, 0.9583, 0.8490, 0.9056, 1.0]
plot_bar_x(precision_val, 'red', 'Precision', 'Precision of Classifiers')
```

```python
falsepositive_rate= [0.0111,0.0111, 0.0222, 0.0888, 0.0555, 0.0 ]
plot_bar_x(falsepositive_rate, 'black', 'FP rate', 'False positive rate of Classifiers')
```
